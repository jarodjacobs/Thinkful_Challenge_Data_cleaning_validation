```python
crime = crime[['City', 'Population', 'Robbery', 'Murder and\nnonnegligent\nmanslaughter']]
```

```python
crime.columns = ['city', 'population', 'robbery', 'murder']
```

```python
crime.robbery = crime.robbery.str.replace(',','')
crime.murder = crime.murder.str.replace(',','')
```

```python
crime.robbery = pd.to_numeric(crime.robbery)
crime.murder = pd.to_numeric(crime.murder)
```
